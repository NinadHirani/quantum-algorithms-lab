```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1)
qc.h(0)
qc.measure_all()

qc.draw()
```

```python
from qiskit_aer import AerSimulator
from qiskit import transpile

sim = AerSimulator()
compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1000).result()

counts = result.get_counts()
counts
```
